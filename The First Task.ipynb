```python
#Code to calculate the radius of a circle
import numpy as np  

x = float(input(print("Input the radius of the circle :")))
area = 3.14*x*x
limited_area = round(area, 2) #This part has been written to limit the number od decimal points
print("The area of the circle with radius", x, " is ", limited_area)

#Code to find out the extension of the given file

file = input("Input the name of the file :")
ext = file.split(".")

if (ext[-1] == "py"):
  print ("The extension of the file is 'python'")
elif (ext[-1] == "txt"):
  print ("The extension of the file is 'text'")
elif (ext[-1] == "img"):
  print ("The extension of the file is 'image'")
elif (ext[-1] == "docx" or ext[-1] == "doc" ):
  print ("The extension of the file is 'document'")
elif (ext[-1] == "bat"):
  print ("The extension of the file is 'batch file'")
elif (ext[-1] == "ppt"):
  print ("The extension of the file is 'powerpoint'")
else :
  print ("The extension of the file is : " + repr(ext[-1]))



```
