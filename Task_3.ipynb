```python
n_terms = int(input(" Input the number of values you would like to see of the Fibonacci sequence :"))

n1, n2 = 0, 1
count = 0

if n_terms <= 0:
    print("Please enter a positive integer")
elif n_terms == 1:
    print("Fibonacci Sequence upto ", n_terms, ":")
    print(n1)
else :
    print ("Fibonacci Sequence")
    while count < n_terms:
        print(n1)
        n = n1 + n2
        n1 = n2
        n2 = n
        count += 1
```
