```python
Input_List = []
Output_List = []

n_terms = int(input("Input how many integers you would like to add to the list :"))


for index in range(0,n_terms):
    num = int(input("Input value number " + str(index + 1) + ":"))
    Input_List.append(num)
    if (Input_List[index] >= 0):
        Output_List.append(Input_List[index])

print("The positive values stored in the Ouput List are :", Output_List)
```
